```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Filter the necessary columns for classification
features = ['koi_period', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_srad', 'koi_steff', 'koi_slogg'] # example feature columns
target = 'koi_disposition'
```

```python
# Remove rows with missing values in important columns
df_cleaned = data[features + [target]].dropna()
```

```python
# 4. Encode the target variable (CONFIRMED = 1, FALSE POSITIVE = 0)
df_cleaned[target] = df_cleaned[target].map({'CONFIRMED': 1, 'FALSE POSITIVE': 0, 'CANDIDATE': -1})
```

```python
# Separate labeled (CONFIRMED and FALSE POSITIVE) and unlabeled (CANDIDATE) data
labeled_data = df_cleaned[df_cleaned[target] != -1]
candidate_data = df_cleaned[df_cleaned[target] == -1]
```

```python
# Features (X) and target (y)
X = labeled_data[features]
y = labeled_data[target]
```

```python
# 5. Split the labeled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 6. Train a Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
# 7. Evaluate the classifier
y_pred = clf.predict(X_test)
print("Accuracy on test data:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# 8. Predict for 'CANDIDATE' exoplanets
X_candidate = candidate_data[features]
candidate_predictions = clf.predict(X_candidate)
```

```python
candidate_data.loc[:, 'prediction'] = candidate_predictions
candidate_data.loc[:, 'prediction'] = candidate_data['prediction'].map({1: 'CONFIRMED', 0: 'FALSE POSITIVE'})
```
